Vậy đối với các quốc gia đang có nền kinh tế không ổn định như nhóm **CPI cao, GDP thấp** và **CPI thấp, GDP thấp**. Các tỷ phú đã đầu tư trong lĩnh vực gì để có thể tạo được khối tài sản khổng lồ như vậy?

```python
# Nhóm kinh tế thấp
low_country = ['Egypt', 'Ukraine', 'Uzbekistan','Nigeria', 'Morocco', 'Greece', 'Cyprus ']
df_low_country = df[df['country'].isin(low_country)]
top_categories = df_low_country['industries'].value_counts().sort_values(ascending=True)

fig = px.bar(x=top_categories.values, y=top_categories.index, orientation='h',color_discrete_sequence=px.colors.qualitative.Set3)
fig.update_layout(title="Industries with the Low Economic Country's Billionaires",
                  xaxis_title='Number of Billionaires',
                  yaxis_title='Industry')
fig.show()
```

##### **Billionaires Segmentation**

Ta sẽ phân nhóm các tỷ phú theo 3 mức độ tài sản:
- **Global Titans:** nhóm tỷ phú có tài sản trên 40 tỷ đô, là nhóm cấp độ cao nhất, gồm những tỷ phú có tầm ảnh hưởng toàn cầu và chi phối nhiều lĩnh vực kinh tế.
- **Industry Leaders:** nhóm có tài sản 20-40 tỷ đô, cấp độ này bao gồm những tỷ phú có ảnh hưởng lớn trong các ngành nghề hoặc lĩnh vực kinh doanh cụ thể.
- **Emerging Magnates:** nhóm có tài sản dưới 20 tỷ đô, đây là nhóm tỷ phú mới nổi, có tiềm năng phát triển lớn và đang nổi bật trong các ngành nghề đặc biệt.

```python
df_emerging_magnates = df[df['finalWorth'] <= 20]
df_industry_leaders = df[(df['finalWorth'] > 20) & (df['finalWorth'] <= 40)]
df_global_titans = df[df['finalWorth'] > 40]
```

```python
dfs = [df_emerging_magnates, df_industry_leaders, df_global_titans]
titles = ['Emerging Magnates', 'Industry Leaders', 'Global Titans']

# Tạo subplots
fig = make_subplots(rows=1, cols=3, subplot_titles=titles)
fig = make_subplots(rows=1, cols=3, subplot_titles=titles, column_widths=[0.33, 0.33, 0.33], specs=[[{'type': 'bar'}, {'type': 'bar'}, {'type': 'bar'}]])

colors = ['rgba(255, 0, 0, 0.6)', 'rgba(0, 255, 0, 0.6)', 'rgba(0, 0, 255, 0.6)']

for i, (df, title, color) in enumerate(zip(dfs, titles, colors), start=1):
    # Đếm số lượng các tỷ phú theo industries
    top_categories = df['industries'].value_counts().sort_values(ascending=True)


    # Vẽ bar chart
    fig.add_trace(go.Bar(x=top_categories.values, y=top_categories.index, orientation='h', marker=dict(color=color)), row=1, col=i)

    fig.update_xaxes(title_text="Number of Billionaires", row=1, col=i)
    fig.update_yaxes(title_text="Category", row=1, col=i)

fig.update_layout(title='Categories of Billionaires Segmentation')

# Hiển thị biểu đồ
fig.show()
```

Ta thấy một điểm khác biệt lớn ở nhóm 'Emerging Magnates', 'Industry Leaders' là số các tỷ phú tập trung trong lĩnh vực "Finance & Investments" nhưng đây đã là nhóm ngành phát triển khá lâu, không có nhiều đột phá và cải tiến nên khó giúp các tỷ phú trong 2 nhóm này có cơ hội đột phá nổi bật.

Trong khi đó ở nhóm Global Titans, các tỷ phú lại là những người tiên phong trong lĩnh vực Technology, nhóm ngành trẻ và hiện đang là xu hướng phát triển trong tương lai nên các tỷ phú đạt được thành công vô cùng nổi trội.
